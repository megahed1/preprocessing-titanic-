```python
df_filled = df.fillna({                   # to replace null (missing values)
    'Age': df['Age'].mean(),
    'Embarked': df['Embarked'].mode()[0]
})

print("\nData with Filled Missing Values:")
print(df_filled)
```

```python
df_filled = df.dropna()           # Removes rows with any missing value
print("\nData with Dropped Missing Values:")
print(df_filled)
```

```python
# remove cabin column

df_filled = df_filled.drop('Cabin', axis=1)
print("\nData with Dropped 'Cabin' Column:")
print(df_filled)
```

```python
 # to find percentage of missing value
df_filled.isnull().sum()/df_filled.shape[0]*100
```

```python
df.drop_duplicates(subset='Name', keep='first')   # to remove duplicate value
```

```python
#creating a new column 'age group'
df_filled['Age Group'] = pd.cut(df_filled['Age'], bins=[0, 25, 30, 100], labels=['Youth', 'Young Adult', 'Adult'])
print(df_filled)
```

```python
# outliers : values out of range
df_filled['Fare'].describe()
```

```python
fig = px.box(df_filled, x='Fare')  # to find outliers
fig.show()
```

```python
q1 = df_filled['Fare'].quantile(0.25)
q3 = df_filled['Fare'].quantile(0.75)
iqr = q3 - q1

outliers = df_filled[(df_filled['Fare'] < q1 - 1.5 * iqr) | (df_filled['Fare'] > q3 + 1.5 * iqr)]
print(outliers)
#Removing outliers #Fixed typo here
df_no_outliers = df_filled[~df_filled.isin(outliers)].dropna()
print("\nData with outliers removed")
print(df_no_outliers)
```

```python
fig = px.box(df_no_outliers, x='Fare') # no outlires
fig.show()
```

```python
df_no_outliers.drop('Name', axis=1, inplace=True)
df_no_outliers.drop('Ticket', axis=1, inplace=True)

print(df_no_outliers)
```

```python
df_no_outliers['Age Group'] = df_no_outliers['Age Group'].astype(str)
df_no_outliers['Embarked'] = df_no_outliers['Embarked'].astype(object)
df_no_outliers['Sex'] = df_no_outliers['Sex'].astype(object)
print(df_no_outliers.dtypes)
```

```python
#one hot encoded
df_encoded = pd.get_dummies(df_no_outliers, columns=['Sex','Embarked','Age Group']).astype(int)#, prefix=['Sex', 'Embarked', 'Age Group']) # Add prefixes for clarity
print("\nEncoded Data:")
print(df_encoded.head())
```
